## 2. Conditional GAN-GPãƒ¢ãƒ‡ãƒ«ã®å®Ÿè£…

```
class Discriminator(nn.Module):
    def __init__(self, class_num):
        super().__init__()
        self.class_num = class_num
        self.conv1 = nn.Conv2d(in_channels=CHANNELS + self.class_num, out_channels=64, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 64 -> 32
        self.conv2 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 32 -> 16
        self.conv3 = nn.Conv2d(in_channels=128, out_channels=256, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 16 -> 8
        self.conv4 = nn.Conv2d(in_channels=256, out_channels=512, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 8 -> 4
        self.conv5 = nn.Conv2d(in_channels=512, out_channels=1, kernel_size=(4, 4), stride=1, padding=0, bias=False) # 4 -> 1

        self.leaky_relu = nn.LeakyReLU(negative_slope=0.2)
        self.dropout = nn.Dropout(p=0.3)
        self.flatten = nn.Flatten()

    def forward(self, x, labels):
        x = torch.cat((x, labels), dim=1) # ç”»åƒã¨ãƒ©ãƒ™ãƒ«ï¼ˆãƒãƒƒãƒã‚µã‚¤ã‚º, ã‚¯ãƒ©ã‚¹æ•°, IMAGE_SIZE, IMAGE_SIZEï¼‰ã®çµåˆ
        x = self.conv1(x)
        x = self.leaky_relu(x)
        x = self.conv2(x)
        x = self.leaky_relu(x)
        x = self.dropout(x)
        x = self.conv3(x)
        x = self.leaky_relu(x)
        x = self.dropout(x)
        x = self.conv4(x)
        x = self.leaky_relu(x)
        x = self.dropout(x)
        x = self.conv5(x)
        x = self.flatten(x)
        
        return x

class Generator(nn.Module):
    def __init__(self, latent_dim, class_num):
        super().__init__()
        self.latent_dim = latent_dim
        self.class_num = class_num

        self.deconv1 = nn.ConvTranspose2d(in_channels=self.latent_dim + self.class_num, out_channels=512, kernel_size=(4, 4), stride=1, padding=0, bias=False) # 1 -> 4
        self.deconv2 = nn.ConvTranspose2d(in_channels=512, out_channels=256, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 4 -> 8
        self.deconv3 = nn.ConvTranspose2d(in_channels=256, out_channels=128, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 8 -> 16
        self.deconv4 = nn.ConvTranspose2d(in_channels=128, out_channels=64, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 16 -> 32
        self.deconv5 = nn.ConvTranspose2d(in_channels=64, out_channels=CHANNELS, kernel_size=(4, 4), stride=2, padding=1, bias=False) # 32 -> 64

        self.leaky_relu = nn.LeakyReLU(negative_slope=0.2)
        self.batch_norm1 = nn.BatchNorm2d(num_features=512, momentum=0.1) # æ³¨ï¼š tfã¨momentumã®å®šç¾©ãŒç•°ãªã‚‹(torch 2.1.0+cu121)
        self.batch_norm2 = nn.BatchNorm2d(num_features=256, momentum=0.1) # æ³¨ï¼š tfã¨momentumã®å®šç¾©ãŒç•°ãªã‚‹(torch 2.1.0+cu121)
        self.batch_norm3 = nn.BatchNorm2d(num_features=128, momentum=0.1) # æ³¨ï¼š tfã¨momentumã®å®šç¾©ãŒç•°ãªã‚‹(torch 2.1.0+cu121)
        self.batch_norm4 = nn.BatchNorm2d(num_features=64, momentum=0.1) # æ³¨ï¼š tfã¨momentumã®å®šç¾©ãŒç•°ãªã‚‹(torch 2.1.0+cu121)
        self.tanh = nn.Tanh()

    def forward(self, x, label_onehot):
        x = torch.cat((x, label_onehot), dim=1) # ãƒ©ãƒ³ãƒ€ãƒ ãƒã‚¤ã‚ºã¨ãƒ©ãƒ™ãƒ«(one-hot)ã®çµåˆ
        x = x.view(x.size(0), self.latent_dim + self.class_num, 1, 1)
        x = self.deconv1(x)
        x = self.batch_norm1(x)
        x = self.leaky_relu(x)
        x = self.deconv2(x)
        x = self.batch_norm2(x)
        x = self.leaky_relu(x)
        x = self.deconv3(x)
        x = self.batch_norm3(x)
        x = self.leaky_relu(x)
        x = self.deconv4(x)
        x = self.batch_norm4(x)
        x = self.leaky_relu(x)
        x = self.deconv5(x)
        x = self.tanh(x)
        
        return x


class ConditionalWGAN(nn.Module):
    def __init__(self, latent_dim, device, gp_weight, d_steps, class_num):
        super().__init__()
        self.latent_dim = latent_dim
        self.class_num = class_num
        self.D = Discriminator(class_num=self.class_num)
        self.G = Generator(latent_dim=self.latent_dim, class_num=self.class_num)
        self.device = device
        self.d_steps = d_steps
        self.gp_weight = gp_weight
        self.d_opt = torch.optim.Adam(self.D.parameters(), lr=LEARNING_RATE, betas=(ADAM_BETA_1, ADAM_BETA_2))
        self.g_opt = torch.optim.Adam(self.G.parameters(), lr=LEARNING_RATE, betas=(ADAM_BETA_1, ADAM_BETA_2))

    def one_hot_encode(self, labels):
        one_hot_labels = nn.functional.one_hot(labels, num_classes=self.class_num).float()
        return one_hot_labels

    def gradient_penalty(self, batch_size, real_images, fake_images, image_one_hot_labels):
        """
        1. è£œå®Œç”»åƒã‚’ç”Ÿæˆ: interpolated = real_images + alpha * (fake_images - real_images)
        2. interpolated.requires_grad_(True): å‹¾é…è¨ˆç®—ã®ãŸã‚ï¼Œinterpolatedãƒ†ãƒ³ã‚½ãƒ«ã«å‹¾é…ã‚’è¨­å®šï¼
        3. å‹¾é…è¨ˆç®—: torch.autograd.gradã‚’ä½¿ç”¨ã—ã¦ï¼Œpredã«å¯¾ã™ã‚‹interpolatedã®å‹¾é…ã‚’è¨ˆç®—ï¼
        4. å‹¾é…ãƒšãƒŠãƒ«ãƒ†ã‚£è¨ˆç®—: å‹¾é…ãƒãƒ«ãƒ ã®L2ãƒãƒ«ãƒ ã‚’è¨ˆç®—ã—ï¼Œç›®æ¨™ã®1ã¨ã®å·®ã‚’2ä¹—ã—ã¦å¹³å‡åŒ–ï¼
        """
        alpha = torch.rand(size=(batch_size, 1, 1, 1), device=self.device)
        diff = fake_images - real_images
        interpolated = real_images + alpha * diff
        interpolated.requires_grad_(True)

        pred = self.D(interpolated, image_one_hot_labels)
        grads = torch.autograd.grad(outputs=pred,
                                    inputs=interpolated,
                                    grad_outputs=torch.ones_like(pred, device=self.device),
                                    create_graph=True,
                                    retain_graph=True,
                                    only_inputs=True)[0]
        norm = torch.linalg.matrix_norm(grads, ord='fro', dim=(-2, -1))
        gp = ((norm - 1) ** 2).mean()

        return gp

    def train_one_step(self, real_images, label_int):
        # å…¥åŠ›ç”¨ã«ãƒ©ãƒ™ãƒ«ã‚’æˆå½¢
        one_hot_labels = self.one_hot_encode(label_int)
        image_one_hot_labels = one_hot_labels[:, :, None, None] # [batch_size, CLASSES, IMAGE_SIZE, IMAGE_SIZE]
        image_one_hot_labels = image_one_hot_labels.repeat(1, 1, IMAGE_SIZE, IMAGE_SIZE)
        # æ½œåœ¨ç©ºé–“å†…ã®ãƒ©ãƒ³ãƒ€ãƒ ãªç‚¹ã‚’ã‚µãƒ³ãƒ—ãƒ«ã™ã‚‹
        batch_size = real_images.size(0)
        # -------------------------
        # è­˜åˆ¥å™¨Dã®å­¦ç¿’
        # -------------------------
        self.D.train()
        for _ in range(self.d_steps):
            self.d_opt.zero_grad()
            random_latent_vectors = torch.randn(size=(batch_size, self.latent_dim), device=self.device)
            # ç”Ÿæˆå™¨ã‹ã‚‰å½ç”»åƒã‚’ç”Ÿæˆï¼ˆå­¦ç¿’ã—ãªã„ã®ã§å‹¾é…ã¯è¨ˆç®—ã—ãªã„ï¼‰
            with torch.no_grad():
                generated_images = self.G(random_latent_vectors, one_hot_labels)

            # Discriminatorã®äºˆæ¸¬
            real_predictions = self.D(real_images, image_one_hot_labels)
            fake_predictions = self.D(generated_images, image_one_hot_labels)

            # Wassersteinæå¤±ã®è¨ˆç®—
            d_wass_loss = -(real_predictions.mean() - fake_predictions.mean())
            # å‹¾é…ãƒšãƒŠãƒ«ãƒ†ã‚£ã®è¨ˆç®—
            d_gp = self.gradient_penalty(batch_size, real_images, generated_images, image_one_hot_labels)
            d_loss = d_wass_loss + self.gp_weight * d_gp

            d_loss.backward()
            self.d_opt.step()

        # ----------------------
        # ç”Ÿæˆå™¨Gã®ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°
        # ----------------------
        self.G.train()
        self.g_opt.zero_grad()
        random_latent_vectors = torch.randn(size=(batch_size, self.latent_dim), device=self.device)
        # ç”Ÿæˆå™¨ã‹ã‚‰å½ç”»åƒã‚’ç”Ÿæˆ
        generated_images = self.G(random_latent_vectors, one_hot_labels)
        fake_predictions = self.D(generated_images, image_one_hot_labels)  # ä»¥ä¸‹ã§ã¯g_optimzerã‚’ä½¿ã†ã®ã§ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¯æ›´æ–°ã•ã‚Œãªã„

        # Wassersteinæå¤±ã®è¨ˆç®—
        g_loss = - fake_predictions.mean()

        g_loss.backward()
        self.g_opt.step()

        # ãƒ¡ãƒˆãƒªã‚¯ã‚¹ã®è¨˜éŒ²
        metrics = {
            "d_loss": d_loss.item(),
            "d_wass_loss": d_wass_loss.item(),
            "d_gp": d_gp.item(),
            "g_loss": g_loss.item(),
        }

        return metrics

    def forward(self):
        pass
```

## 3. CGAN-GPã®å­¦ç¿’

```
def save_model(model, path):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    model.eval()
    torch.save(model.state_dict(), path)

def load_model(model, path, device):
    model.load_state_dict(torch.load(path, map_location=device))
    model = model.to(device)
    model.eval()
    print(f"Model loaded from {path}")
    return model
```

```
metrics_per_epoch = {
    "d_loss": [],
    "d_wass_loss": [],
    "d_gp": [],
    "g_loss": [],
}
cgan = cgan.to(device)

# ä¿å­˜ã™ã‚‹ã‚¨ãƒãƒƒã‚¯ã‚’ãƒªã‚¹ãƒˆã§æŒ‡å®š
save_epochs = [1, 50, 100, 150, 200]

for epoch in range(EPOCHS):
    epoch_metrics = {metric: 0 for metric in metrics_per_epoch}
    batch_count = 0

    for i, data in enumerate(trainloader, 0):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        # ãƒŸãƒ‹ãƒãƒƒãƒå­¦ç¿’
        metrics = cgan.train_one_step(inputs, labels)

        for key in epoch_metrics:
            epoch_metrics[key] += metrics[key]

        batch_count += 1

    # ã‚¨ãƒãƒƒã‚¯ã®ãƒ¡ãƒˆãƒªã‚¯ã‚¹å¹³å‡ã‚’è¨ˆç®—ã—ã¦ä¿å­˜
    for key in epoch_metrics:
        epoch_metrics[key] /= batch_count
        metrics_per_epoch[key].append(epoch_metrics[key])

    # ãƒ¡ãƒˆãƒªã‚¯ã‚¹ã®å‡ºåŠ›
    print("Epoch {}:".format(epoch + 1), epoch_metrics)

    # æŒ‡å®šã—ãŸã‚¨ãƒãƒƒã‚¯ã®ã¿ä¿å­˜
    if (epoch + 1) in save_epochs:

        # ãƒ¢ãƒ‡ãƒ«ã‚’ä¿å­˜
        save_path = f"./models/08_cgan_model_epoch_{epoch + 1}.pth"
        save_model(cgan, save_path)
        print(f"Model saved to {save_path}")


save_path = f"./models/08_cgan_model_epoch_{EPOCHS}.pth"
save_model(cgan, save_path)

metrics_save_path = f"./metrics/08_cgan_metrics.json"
with open(metrics_save_path, "w") as f:
    for metrics in metrics_per_epoch:
        f.write(json.dumps(metrics) + "\n")  # å„ã‚¨ãƒãƒƒã‚¯ã®ãƒ¡ãƒˆãƒªã‚¯ã‚¹ã‚’1è¡Œãšã¤ä¿å­˜

print(f"Metrics saved to {metrics_save_path}")
```
